```python
import pandas as pd 
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
# ATOM's wrapper for binary and multiclass classification tasks.
from atom import ATOMClassifier
```

```python
# Read, drop , decrease decimals to 4, clean whitespaces.
df = pd.read_csv('Test-set_2 Anonymized.csv') 
df.drop(['X__2'],axis=1,inplace=True)
df = round(df, 4)
df.replace(' ', '', regex=True)
```

```python
# Replace V1_0 with 0 , replace V1_1 with 1 
df["X__1"].replace({"V1_0": "0", "V1_1": "1"}, inplace=True)
# Convert datatype to integer
df["X__1"] = df["X__1"].astype('int')
# Define Independent columns and dependent (target) column.
y = df["X__1"]
X = df.iloc[:,1:] 
```

```python
df.head(3)
```

```python
# Shuffle df before trainings.
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df
```

```python
# Function and code for splitting data into train and test, 80% of data used for train.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=41, train_size = .8)
```

```python
# List contains the selected models from first notebook.
listformodels = ['AdaB',
 'CatB',
 'hGBM',
 'XGB']
```

```python
# Inıtıalize atom with defined train and test sets, set verbose = 2 to see more informations when models running.
atom = ATOMClassifier(X_train, X_test, y_train, y_test, verbose=0,                    
                      n_jobs=-1,
                      random_state=48)

atom.impute(strat_num="median")
atom.encode(max_onehot=5)
atom.scale(strategy="robust")
```

# Correlations

```python
# Correlations between some of the columns.
atom.plot_correlation()
```

# Accuracy is not a good metric to use when you have class imbalance.
# The F1 score: Combining Precision and Recall
Precision and Recall are the two building blocks of the F1 score. The goal of the F1 score is to combine the precision and recall metrics into a single metric. At the same time, the F1 score has been designed to work well on imbalanced data.

```python
listformodels = ['AdaB',
 'CatB',
 'hGBM',
 'XGB']

# Run base models 
for i in listformodels:
        atom.run(models= i+"base" ,metric=("F1"))

# Cross Validate Models with 3 fold.
atom.hGBMbase.calibrate(cv=3)
atom.CatBbase.calibrate(cv=3)
atom.XGBbase.calibrate(cv=3)
atom.AdaBbase.calibrate(cv=3)
```

# Runs 6 different oversampling methods on 4 models with cross-validate models

```python
# List contains different oversampling methods.
listforoversamplers = ['randomoversampler', 'smote', 'borderlinesmote', 'kmeanssmote', 'svmsmote',]

for i in listforoversamplers:
    # Came back to master branch
    atom.branch="master"
    # Create branch from master branch for every oversampler in listforoversamplers.
    atom.branch = f"{i}"+ "_from_master"
    # Get current branchs name for naming the model
    a = atom.branch.name
    # Apply oversampling method to branch. Datas will be different on every branch.
    atom.balance(strategy=i)
    
    # Run every model for every branch.    
    for i in listformodels:
        # Name models with branch names for understanding which oversampling method is better for data.
        atom.run(models= i + f"{a}",metric=("F1"))
        
```

```python
# Get fitted models 
trainedmodels = atom.models
```

```python
listforcv=[]
for i in trainedmodels[4:]:
    a = 'atom'+"."+f"{i}"
    listforcv.append(a)    
```

```python
# Model names for cross validating.
listforcvv = [atom.AdaBrandomoversampler,atom.CatBrandomoversampler,atom.hGBMrandomoversampler,atom.XGBrandomoversampler,
             atom.AdaBsmote,atom.CatBsmote,atom.hGBMsmote,atom.XGBsmote,atom.AdaBborderlinesmote,
             atom.CatBborderlinesmote,atom.hGBMborderlinesmote,atom.XGBborderlinesmote,atom.AdaBkmeanssmote,
              atom.hGBMkmeanssmote,atom.XGBkmeanssmote,atom.AdaBsvmsmote,atom.CatBsvmsmote,atom.hGBMsvmsmote,
             atom.XGBsvmsmote]
```

```python
# Cross Validate All of Them with 3 fold.
for i in listforcvv:
    i.calibrate(cv=3)
```

# Created 20 features for CatBoost with Deep Feature Synthesis 

```python
# Change current branch to branch that I've used for svmsmote.
atom.branch='svmsmote'

# Give name to created branch.
atom.branch ="_svmsmote"+"dfs" 

# Get current branch name 
a=atom.branch.name


# Create new combinations of existing features to capture the non-linear relations between the original features.
atom.feature_generation(
        
        strategy="dfs", # Deep Feature Synthesis 
        
        n_features=20, # Maximum number of newly generated features to add to the dataset.
        
        operators=["add","mul","sub"]  # You can add or subtract different data transformations here.
        
)

atom.run(models = "CatB"+"dfs" + f"{a}", metric=("F1"))
```

# New CatBoost Pipeline 

```python
atom.CatBdfs_svmsmotedfs.plot_pipeline()
```

# Feature Selection with RFE, choosing optimal number of features.

```python
# Select 45, 55, 65 features and finds best number as column count to choose.
for i in range(45,66,10):
    atom.branch = "_svmsmotedfs"
    atom.branch =  "_svmsmotedfs" + "_fsel" + f"{i}" 
    a=atom.branch.name
    atom.feature_selection("RFE", solver="RF", n_features=i )
    atom.run(models = "CatB"+"dfs" + f"{a}", metric=("F1"))
```

# Choosing 55 features improved Recall 

```python
atom.evaluate().sort_values("f1",ascending=False).loc[['CatBdfs_svmsmotedfs_fsel55','CatBdfs_svmsmotedfs']]
```

# A true positive is an outcome where the model correctly predicts the positive class. Similarly, a true negative is an outcome where the model correctly predicts the negative class.

# A false positive is an outcome where the model incorrectly predicts the positive class. And a false negative is an outcome where the model incorrectly predicts the negative class.

```python
atom.plot_confusion_matrix(models=["hGBMHyper_finall","CatBHyper"])
```

# Precision: It is the accuracy of positive predictions.
# Recall: It is the ratio of positive instance that are correctly detected. It is also called sensitivity.

# 1- A model with high recall succeeds well in finding all the positive cases in the data, even though they may also wrongly identify some negative cases as positive cases.
# 2- A model with low recall is not able to find all (or a large part) of the positive cases in the data.

# 3- If you increase precision, it will reduce recall, and vice versa. This is called the precision/recall tradeoff.

```python
atom.plot_prc(models=["hGBMHyper_finall","CatBHyper"])
```

```python
atom.plot_roc(models=["hGBMHyper_finall","CatBHyper"])
```
